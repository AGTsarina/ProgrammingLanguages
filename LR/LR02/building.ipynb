```python
# Загрузка исходных данных
import geopandas as gpd

df = gpd.read_file('01_Новая Голландия.geojson')
#  Представлена в виде однотабличной базы данных
# full_id - уникальный номер здания
# osm_id - уникальный номер на карте OSM
# osm_type - тип географического объекта
# building - тип здания
# Area - площадь здания
# geometry - один или несколько многоугольников, упакованы в тип "набор многоугольников" MULTIPOLYGON
df
```
