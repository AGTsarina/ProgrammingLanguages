```python
from math import sin
def f(x):
    return 5 * sin(x) - x + 20
```

```python
x = []
y = []
for i in range(2000):
    x.append(i * 0.01)
    y.append(f(i * 0.01))
```

```python
import matplotlib.pyplot as plt
plt.plot(x, y)
```

```python
xIn, yIn, xOut, yOut = [], [], [], []
with open('res.txt', 'r') as f:
    lines = f.readlines()
    for line in lines:
        words = line.split()
        x, y = float(words[0]), float(words[1])
        if words[2].startswith('In'):
            xIn.append(x)
            yIn.append(y)
        else:
            xOut.append(x)
            yOut.append(y)
plt.scatter(xIn, yIn)
plt.scatter(xOut, yOut)
```
